A proper look at the dataframe shows many Series have no values. There is also no input for 2021. So I will have to drop the affected series and rows

Now pivoting the dataframe to make it suitable for analyses.

```python
New_df= New_df.loc[["Labor force, female (% of total labor force)", "Lifetime risk of maternal death (%)", 
                    "Maternal mortality ratio (modeled estimate, per 100,000 live births)", "Number of maternal deaths" ]]
New_df.head()
```

```python
print(New_df)
```

There is almost no input from 1960 to 2011. And no input in 2021 too. These columns will be dropped

```python
New_df= New_df.drop(['1960', '1970',], axis = 1)
```

```python
New_df = New_df.reset_index()
```

```python
New_df = pd.melt(New_df, id_vars =["Series Name"], var_name = "Year", value_name ="Rate")
New_df.head()
```

```python
New_df["Rate"].min()
```

```python
New_df["Rate"].max()
```

```python
list(New_df["Series Name"].unique())
```

```python
fig, ax =plt.subplots(figsize= (30,20), dpi = 80)
sns.set_theme(style="darkgrid")

sns.lineplot(y= "Rate", x= "Year", data = New_df, hue = "Series Name", linewidth =5, style = "Series Name", )

ax.set_title("Rates across the years", fontdict = {"fontsize":50})
ax.set_xlabel("Year", fontsize = 50)
ax.set_ylabel("Numbers", fontsize = 50)
ax.tick_params(labelsize = 20)
ax.legend(loc= "upper right", bbox_to_anchor=(1.6,1))
ax.grid(visible = True, which = "minor",  linewidth= 0.25 )
ax.autoscale(enable= True, axis = "both")
```
